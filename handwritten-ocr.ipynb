```python
# ============================================================
#     HANDWRITTEN OCR USING EMNIST LETTERS (CSV VERSION)
# ============================================================


# ---------------------- 1. IMPORT LIBRARIES ----------------------
!pip install protobuf==3.20.* --quiet

import os
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"
os.environ["XLA_FLAGS"] = "--xla_gpu_cuda_data_dir=/usr/lib/cuda/"

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout



# ============================================================
# 2. LOAD EMNIST LETTERS (CSV FORMAT)
# ============================================================

# Paths based on your dataset screenshot
train_path = "/kaggle/input/emnist/emnist-letters-train.csv"
test_path  = "/kaggle/input/emnist/emnist-letters-test.csv"

# Load CSV (no header row)
train = pd.read_csv(train_path, header=None)
test  = pd.read_csv(test_path, header=None)

print("Train shape:", train.shape)
print("Test shape :", test.shape)



# ============================================================
# 3. SPLIT LABELS (y) AND IMAGES (X)
# ============================================================

# Labels are in column 0 and go from 1 to 26 → must convert to 0–25
y_train = train.iloc[:, 0].values - 1
y_test  = test.iloc[:, 0].values - 1

# Image pixels (784 columns)
X_train = train.iloc[:, 1:].values
X_test  = test.iloc[:, 1:].values



# ============================================================
# 4. PREPROCESS IMAGES
# ============================================================

# Normalize 0–255 → 0–1
X_train = X_train / 255.0
X_test  = X_test  / 255.0

# Reshape 784 → 28×28×1
X_train = X_train.reshape(-1, 28, 28, 1)
X_test  = X_test.reshape(-1, 28, 28, 1)

# Fix EMNIST rotation issue (rotate images to correct orientation)
X_train = np.rot90(X_train, k=1, axes=(1, 2))
X_test  = np.rot90(X_test,  k=1, axes=(1, 2))



# ============================================================
# 5. BUILD CNN MODEL (NO WARNINGS)
# ============================================================

model = Sequential([
    
    # Input layer avoids warnings about input_shape in Conv2D
    Input(shape=(28, 28, 1)),

    # Convolution layer 1: learns edges, curves
    Conv2D(32, (3,3), activation='relu'),
    MaxPooling2D((2,2)),

    # Convolution layer 2: learns deeper patterns
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),

    Flatten(),               # Flatten into a single vector
    Dense(128, activation='relu'),
    Dropout(0.4),            # Prevent overfitting

    Dense(26, activation='softmax')  # Output for 26 classes A–Z
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()



# ============================================================
# 6. TRAIN THE MODEL
# ============================================================

history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=12,
    batch_size=128,
    verbose=1
)



# ============================================================
# 7. EVALUATE ACCURACY
# ============================================================

loss, accuracy = model.evaluate(X_test, y_test)
print("Final Test Accuracy:", accuracy)



# ============================================================
# 8. SAMPLE PREDICTION
# ============================================================

idx = 50  # choose any test index

plt.imshow(X_test[idx].squeeze(), cmap='gray')
plt.title("Sample Test Image")
plt.show()

pred = model.predict(X_test[idx].reshape(1,28,28,1))
predicted_label = np.argmax(pred)

print("Predicted Letter Label:", predicted_label)



# ============================================================
# 9. SAVE MODEL (PERMANENT)
# ============================================================

model.save("/kaggle/working/ocr_emnist_model.keras")
print("Model saved to /kaggle/working/ocr_emnist_model.keras")


# ============================================================
# 10. SAVE ALL PREDICTIONS TO CSV (PERMANENT)
# ============================================================

preds_all = np.argmax(model.predict(X_test), axis=1)

results_df = pd.DataFrame({
    "True_Label": y_test,
    "Predicted_Label": preds_all
})

results_df.to_csv("/kaggle/working/ocr_predictions.csv", index=False)
print("Saved predictions → /kaggle/working/ocr_predictions.csv")



# ============================================================
# 11. SAVE MISCLASSIFIED EXAMPLES (PERMANENT)
# ============================================================

wrong_idx = np.where(preds_all != y_test)[0]

mis_df = pd.DataFrame({
    "Index": wrong_idx,
    "True_Label": y_test[wrong_idx],
    "Predicted_Label": preds_all[wrong_idx]
})

mis_df.to_csv("/kaggle/working/misclassified.csv", index=False)
print("Saved misclassified → /kaggle/working/misclassified.csv")



# ---------------------- END OF PROJECT ----------------------
```
